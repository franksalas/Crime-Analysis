```python
import pandas as pd
import numpy as np
```

```python
data =  'crime_data_clean/crime_final_clean10_17.csv'
```

```python
df = pd.read_csv(data)
```

```python
df.info()
```

```python
# date
df.Date = pd.to_datetime(df.Date)

df = df.set_index('Date').sort_index(ascending=True)
```

```python
df.head()
```

## Select Beats
- create list with select police beats where selected stadiums are located
- filter dataframe with those beats

```python
selected_beats = ['1A10','10H30','10H50','10H80','15E40' ]
```

```python
sel_df = df[df['Beat'].isin(selected_beats)]
```

```python
sel_df.info()
```

## Inspect BlockRange

```python
len(sel_df[sel_df.BlockRange.isnull()])
```

```python
sel_df[sel_df.BlockRange.isnull()].head()
```

```python
sel_df.Beat.value_counts(dropna=False)
```

```python
sel_df = sel_df.dropna(subset=['BlockRange'])
```

```python
len(sel_df[sel_df.BlockRange.isnull()])
```

```python
sel_df.Beat.value_counts(dropna=False)
```

----

# Save Final Clean Data

```python
sel_df.to_csv('crime_data_clean/beats_filter_10_17.csv')
```

```python
crime_10H50_10_17.to_csv('crime_data_clean/crime_10H50_10_17.csv')
crime_1A10_10_17.to_csv('crime_data_clean/crime_1A10_10_17.csv')
crime_10H80_10_17.to_csv('crime_data_clean/crime_10H80_10_17.csv')
crime_15E40_10_17.to_csv('crime_data_clean/crime_15E40_10_17.csv')
crime_10H30_10_17.to_csv('crime_data_clean/crime_10H30_10_17.csv')
```

```python
path = 'crime_data_clean/df_15E40FINAL.csv'
df = pd.read_csv(path)
df.drop(['Unnamed: 0'],axis=1,inplace=True)
df.head(2)
```

```python
stadium = stadium_l[::-1]



subset = df[['lat','lon']]
tp = [tuple(x) for x in subset.values]


dd = []
for i,p in tp:
    c =  (p,i)
    start = lonlat(*stadium)
    end = lonlat(*c)
    dis = distance(start,end).miles
    #dis =  geopy.distance.distance(stadium, c).miles
    dd.append(dis)
    
dist = pd.Series(dd,name='dist_stadium')
```

```python
df = df.join(dist)
```

```python
df.head(2)
```

```python
one_mile =  df.dist_stadium < 1.0
one = df[one_mile]
```

```python
m = folium.Map([29.7281, -95.2298],     tiles='Stamen Toner',
zoom_start=11)

# convert to (n, 2) nd-array format for heatmap
stationArr = one[['lat', 'lon']].as_matrix()
folium.Marker(stadium_l, popup='<i>Texams Stadium</i>').add_to(m)


# add beat geojson file
folium.GeoJson(beat1,
    name='geojson'
).add_to(m)



folium.Circle(
    radius=1609,
    location=stadium_l,
    popup='The Waterfront',
    color='crimson',
    fill=False,
).add_to(m)

# plot heatmap
m.add_child(plugins.HeatMap(stationArr.tolist(), radius=15))

m
```

# Scores dataframe

```python
games = 'sports_scores/Tex_15E40.csv'
```

```python
df_game = pd.read_csv(games)


## split date and time from column scheduled
s_date = pd.DatetimeIndex(df_game['scheduled'])
df_game['Date'] = s_date.date
df_game['time'] = s_date.time
del df_game['scheduled']

# set date as datetime and index
df_game.Date = pd.to_datetime(df_game.Date)
df_game = df_game.set_index('Date').sort_index(ascending=True)






## win column
df_game['win'] =  np.where(df_game['scoring.home_points'] > df_game['scoring.away_points'],1,0)

# only oposing name and win boolean
df_game = df_game[['win']]

# show

df_game.head()
```

# Days dataframe

```python
days = pd.date_range(start='01/01/2010', end='12/31/2017')
days = pd.DataFrame(days)

days.columns = ['days']
days = days.set_index('days').sort_index(ascending=True)
days.head()
```

# days and games merge

```python
# calendar_crimes = pd.merge(days,date_crimes,  left_index=True, right_index=True, how='left')
# calendar_crimes.head()
calendar_games = pd.merge(days,df_game,  left_index=True, right_index=True, how='left')
calendar_games.head()
```

```python
crime_texans = pd.merge(calendar_games,texans,  left_index=True, right_index=True, how='left')
```

```python
crime_texans.info()
```

```python
def game_feature(df):
    if df.win == 1:
        val = 'Won Game'
    elif df.win == 0:
        val = 'Lost Game'
    else:
        val = 'No Game'
    return val
```

```python
crime_texans['game'] = crime_texans.apply(game_feature,axis=1)
```

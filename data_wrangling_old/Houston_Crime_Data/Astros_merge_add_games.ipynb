# 1A10

```python
path = 'crime_data_clean/df_1A10FINAL.csv'
df = pd.read_csv(path)
df.drop(['Unnamed: 0'],axis=1,inplace=True)
df.head(2)
```

```python
stadium = stadium_l[::-1]



subset = df[['lat','lon']]
tp = [tuple(x) for x in subset.values]


dd = []
for i,p in tp:
    c =  (p,i)
    start = lonlat(*stadium)
    end = lonlat(*c)
    dis = distance(start,end).miles
    #dis =  geopy.distance.distance(stadium, c).miles
    dd.append(dis)
    
dist = pd.Series(dd,name='dist_stadium')
```

```python
df = df.join(dist)
```

```python
df.head(2)
```

```python
one_mile =  df.dist_stadium < 1.0
one = df[one_mile]
```

```python
m = folium.Map([29.7281, -95.2298],     tiles='Stamen Toner',
zoom_start=11)

# convert to (n, 2) nd-array format for heatmap
stationArr = one[['lat', 'lon']].as_matrix()
folium.Marker(stadium_l, popup='<i>Astros Stadium</i>').add_to(m)


# add beat geojson file
folium.GeoJson(beat1,
    name='geojson'
).add_to(m)


# add beat geojson file
folium.GeoJson(beat2,
    name='geojson'
).add_to(m)

# add beat geojson file
folium.GeoJson(beat3,
    name='geojson'
).add_to(m)



folium.Circle(
    radius=1609,
    location=stadium_l,
    popup='The Waterfront',
    color='crimson',
    fill=False,
).add_to(m)

# plot heatmap
m.add_child(plugins.HeatMap(stationArr.tolist(), radius=15))

m
```

# Scores dataframe

```python
games = 'sports_scores/astros_1A10.csv'
```

```python
df_game = pd.read_csv(games,index_col='date').sort_index(ascending=True)


## win column
df_game['win'] =  np.where(df_game['team_score'] > df_game['opp_score'],1,0)

# only oposing name and win boolean
df_game = df_game[['win']]

# show



df_game.head()
```

# Days dataframe

```python
days = pd.date_range(start='01/01/2010', end='12/31/2017')
days = pd.DataFrame(days)

days.columns = ['days']
days = days.set_index('days').sort_index(ascending=True)
days.head()
```

# days and games merge

```python
# calendar_crimes = pd.merge(days,date_crimes,  left_index=True, right_index=True, how='left')
# calendar_crimes.head()
calendar_games = pd.merge(days,df_game,  left_index=True, right_index=True, how='left')
calendar_games.head()
```

```python
crime_astros = pd.merge(calendar_games,astros,  left_index=True, right_index=True, how='left')
```

```python
crime_astros.info()
```

```python
crime_astros.head(2)
```

```python
def game_feature(df):
    if df.win == 1:
        val = 'Won Game'
    elif df.win == 0:
        val = 'Lost Game'
    else:
        val = 'No Game'
    return val
```

```python
crime_astros['game'] = crime_astros.apply(game_feature,axis=1)
```

# 10H80

```python
path = 'crime_data_clean/df_10H80FINAL.csv'
df = pd.read_csv(path)
df.drop(['Unnamed: 0'],axis=1,inplace=True)
df=df.rename(columns = {'full_address':'add'})
df.head(2)
```

```python
stadium = stadium_l[::-1]



subset = df[['lat','lon']]
tp = [tuple(x) for x in subset.values]


dd = []
for i,p in tp:
    c =  (p,i)
    start = lonlat(*stadium)
    end = lonlat(*c)
    dis = distance(start,end).miles
    #dis =  geopy.distance.distance(stadium, c).miles
    dd.append(dis)
    
dist = pd.Series(dd,name='dist_stadium')
```

```python
df = df.join(dist)
```

```python
df.head(2)
```

```python
one_mile =  df.dist_stadium < 1.0
one = df[one_mile]
```

```python
m = folium.Map([29.7281, -95.2298],     tiles='Stamen Toner',
zoom_start=11)

# convert to (n, 2) nd-array format for heatmap
stationArr = one[['lat', 'lon']].as_matrix()
folium.Marker(stadium_l, popup='<i>Rice Stadium</i>').add_to(m)


# add beat geojson file
folium.GeoJson(beat1,
    name='geojson'
).add_to(m)


# add beat geojson file
folium.GeoJson(beat2,
    name='geojson'
).add_to(m)



folium.Circle(
    radius=1609,
    location=stadium_l,
    popup='The Waterfront',
    color='crimson',
    fill=False,
).add_to(m)

# plot heatmap
m.add_child(plugins.HeatMap(stationArr.tolist(), radius=15))

m
```

# Scores dataframe

```python
games = 'sports_scores/Rice_10H80.csv'
```

```python
df_game = pd.read_csv(games,index_col='Date').sort_index(ascending=True)

## win column
df_game['win'] =  np.where(df_game['Pts'] > df_game['Opp'],1,0)

# only oposing name and win boolean
# df = df[['Opponent','win','beat']]
df_game = df_game[['win']]

# show

df_game.head()
```

# Days dataframe

```python
days = pd.date_range(start='01/01/2010', end='12/31/2017')
days = pd.DataFrame(days)

days.columns = ['days']
days = days.set_index('days').sort_index(ascending=True)
days.head()
```

# days and games merge

```python
# calendar_crimes = pd.merge(days,date_crimes,  left_index=True, right_index=True, how='left')
# calendar_crimes.head()
calendar_games = pd.merge(days,df_game,  left_index=True, right_index=True, how='left')
calendar_games.head()
```

```python
crime_ru = pd.merge(calendar_games,cfbRU,  left_index=True, right_index=True, how='left')
```

```python
crime_ru.info()
```

```python
crime_ru.head(2)
```

```python
def game_feature(df):
    if df.win == 1:
        val = 'Won Game'
    elif df.win == 0:
        val = 'Lost Game'
    else:
        val = 'No Game'
    return val
```

```python
crime_ru['game'] = crime_ru.apply(game_feature,axis=1)
```

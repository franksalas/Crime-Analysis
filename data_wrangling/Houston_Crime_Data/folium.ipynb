```python
df['lat_lon'] = df['lat_lon'].str.extract(r"\((.*?)\)", expand=False)
```

```python
ttt = pd.DataFrame(df.lat_lon.str.split(',',1).tolist(), columns=['lat','lon'])
```

```python
ttt.dtypes
```

```python
ttt.lat = ttt.lat.astype('float')
ttt.lon = ttt.lon.astype('float')
```

```python
locations = ttt[['lat', 'lon']]
locationlist = locations.values.tolist()
len(locationlist)
locationlist[7]
```

```python
map = folium.Map(location=[29.9, -95.05], tiles='Stamen Toner',zoom_start=12)
for point in range(0, len(locationlist)):
    folium.Marker(locationlist[point],popup=df.loc[point].full_address).add_to(map)
```

```python
map.save('add.html')
```

```python

import pandas as pd
import numpy as np
```

```python
rice_games = 'data/sports_scores/Rice_10H80.csv'
```

```python
crime_10H80 = 'data/crime_beats/crime_10H80_10_17.csv'
```

```python
df_rice = pd.read_csv(rice_games,index_col='Date')
crime_rice = pd.read_csv(crime_10H80,index_col='Date')
```

```python
df_rice.head()
```

```python
df_rice['win'] =  np.where(df_rice.Pts > df_rice.Opp,1,0)
```

```python
df_rice = df_rice[['Opponent','win']]
```

```python
crime_rice = crime_rice.iloc[2010:]  # data from 2010- 2017
```

```python
crime_rice = crime_rice[['OffenseType','Premise','NumOffenses','Hour','weekday']]
```

## merge

```python
merge = pd.merge(df_rice,crime_rice, how='inner', left_index=True, right_index=True)
```

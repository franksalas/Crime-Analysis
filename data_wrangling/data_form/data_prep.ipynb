```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
plt.style.use('ggplot')
```

```python
path =  'data/final/Rice1017_final.csv'
```

```python
df = pd.read_csv(path,index_col='days',parse_dates=True)
#df = pd.read_csv(path)
df.head()
```

```python
df.info()
```

```python
df.won.value_counts(dropna=False)
```

```python
def game_feature(df):
    if df.won == 1:
        val = 'Won Game'
    elif df.won == 0:
        val = 'Lost Game'
    else:
        val = 'No Game'
    return val
```

```python
df['game'] = df.apply(game_feature,axis=1)
```

```python
df.game.value_counts()
```

```python
df_dy = df[['crime','game']]
```

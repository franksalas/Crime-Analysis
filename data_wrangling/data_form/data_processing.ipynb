## Baseball

```python
path = 'data/final/Astros1017_final.csv'
df = pd.read_csv(path)
df.head()
```

```python
#print(list(df.columns))
df.columns = ['days', 'crime', 'away_team', 'won', 'beat', 'sport']
```

```python
df.head()
```

```python
df['game'] = np.where((df['away_team'].isnull()),0,1)
```

```python
df.game.value_counts()
```

```python
df['days'] = pd.to_datetime(df['days'])
df['days'] = df['days'].dt.date
df = df.set_index('days')
```

```python
df.head()
```

```python
bas_df = df[['crime','game','beat','sport']]
```

```python
bas_df.head()
```

```python
bas_df.to_csv('data/ready/bas.csv')
```

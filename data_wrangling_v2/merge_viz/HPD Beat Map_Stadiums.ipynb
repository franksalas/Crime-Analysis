# plot Stadiums and Beats

```python
nrg_stadium = [29.684722, -95.410833]
rice_stadium = [29.716389, -95.409167]
bbva_stadium = [29.7522, -95.3524]
tedecu_stadium = [29.721944, -95.349167]
toyota_stadium = [29.750833, -95.362222]
mm_stadium = [29.756944, -95.355556]
```

```python
m = folium.Map(Houston,     tiles='Stamen Toner',
zoom_start=12)

# add beat geojson file
folium.GeoJson(beat1,
    name='geojson'
).add_to(m)

folium.Marker(bbva_stadium, popup='<i>BBVA Stadium</i>').add_to(m)
folium.Marker(tedecu_stadium, popup='<i>TDECU Stadium</i>').add_to(m)
folium.Marker(toyota_stadium, popup='<i>Toyota Center</i>').add_to(m)
folium.Marker(mm_stadium, popup='<i>MinuteMaid Park</i>').add_to(m)

folium.Marker(nrg_stadium, popup='<i>NRG Stadium</i>').add_to(m)
folium.Marker(rice_stadium, popup='<i>Rice Stadium</i>').add_to(m)

# plot heatmap
m.save('images/BEAT.html')

m
```

```python
path = 'data/MLB/reg'
```

```python
all_files = glob.glob(os.path.join(path, "*.csv")) 

df_from_each_file = (pd.read_csv(f) for f in all_files)
df   = pd.concat(df_from_each_file, ignore_index=True)
```

```python
df.head()
```

## slice neede cols

```python
mlb = df[['Date','Year','Tm','R','Unnamed: 5','Opp','RA','W/L']]
mlb.head()
```

## Convert yeaer column to string

```python
mlb['Year'] = mlb['Year'].astype(str)
```

## create new col with full date

```python
mlb['full_date'] = mlb['Date']+' '+ mlb['Year']
```

```python
mlb.head()
```

## Remove parenthesis from dates from double  games

```python
mlb[mlb.full_date.str.contains('(', regex=False)]
```

```python
mlb['full_date'] = mlb['full_date'].str.replace(r"\(.*\)"," ")
```

```python
mlb[mlb.full_date.str.contains('(', regex=False)]
```

```python
mlb
```

## Convert date to datetime 

```python
mlb['date'] = pd.to_datetime(mlb['full_date'])

mlb = mlb.set_index('date').sort_index(ascending=True)
```

```python
mlb.head()
```

## Select columns

```python
astros = mlb[['Tm','R','Unnamed: 5','Opp','RA','W/L']]
astros.head()
```

## Rename columns

```python
astros.rename(columns={
                       'Tm': 'team',
                       'R':'team_score',
                       'Unnamed: 5':'home_away',
                      'Opp': 'opposing',
                      'RA':'opp_score',
                      'W/L':'win_lost',
                      }, inplace=True)
```

```python
astros.head()
```

## Change column
- change home_away values to  home or away

```python
astros['home_away'] = np.where(astros['home_away'] =='@', 'AWAY', 'HOME')
```

```python
astros.head()
```

## Find and replace
 - replace 'W-wo' : 'W'
 - replace  'L-wo': 'L'
 - replace 'W &H' : 'W

```python
astros.win_lost.unique()
```

```python
astros['win_lost'] = astros['win_lost'].str.replace('L-wo','L')
astros['win_lost'] = astros['win_lost'].str.replace('W-wo','W')
astros['win_lost'] = astros['win_lost'].str.replace('W &H','W')
```

```python
astros.win_lost.unique()
```

## add col for season

```python
astros['season'] = 'reg'
```
